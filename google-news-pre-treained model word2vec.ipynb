### average word2vec for pre-trained model

we use our custom defined package of the word2vec

```python
from word2vec.Averageword2vec import avg_word_2_vector as avg_vector #this is our predefined module we made it inside vs code of source code AI and then vectorization>word2vec it is tested in test.py
```

```python
import gensim.downloader as api
```

```python
model =api.load("word2vec-google-news-300")
```

```python
sentence="maxhine learning is fun and amazing"
vector=avg_vector(sentence,model)
print("average word2vec vector:",vector)
```

### gender similarity:king,man and woman

```python
print("vector of king:",model["king"])
print("vector of man:",model["man"])
print("vector of woman:",model["woman"])
```

```python
print("similarity between king and queen:",model.similarity("king","queen"))
```

```python
similarity=model.similarity("king","queen")*100
print("percentage of similarity:",round(similarity,2))
```

### Q1: if man is to king then what will be woman is _____?

```python
#vector("man") and vector("king") -> model ----->women ->predication
```

```python
result=model.most_similar(positive=["woman","king"],negative=["man"])
```

```python
answer,score=result[0]
```

```python
print("answer with score:",answer,score)
```

```python
result_new=model.most_similar(positive=["Delhi","USA"],negative=["India"])
```

```python
answer,score=result_new[0]
print("answer with score:",answer,score)
```
